```python
def run_One_Game(team_A,team_B,rating_A,rating_B,df_group_Alphabet) :
    
    win_Rate_A = 1 / (1 + 10**( (rating_B-rating_A) / 400) )
    
    num_Rand = random.random()
    print(f"{team_A}의 승리 확률 : {win_Rate_A*100}")
    print(f"{team_B}의 승리 확률 : {(1-win_Rate_A)*100}")
    
    print(num_Rand)
    
    if num_Rand <= win_Rate_A:
        df_group_Alphabet.loc[df_group_Alphabet['team_Name']==team_A,'win'] += 1
        df_group_Alphabet.loc[df_group_Alphabet['team_Name']==team_A,'point'] += 3
        df_group_Alphabet.loc[df_group_Alphabet['team_Name']==team_B,'lose'] += 1
    else :
        df_group_Alphabet.loc[df_group_Alphabet['team_Name']==team_A,'lose'] += 1
        df_group_Alphabet.loc[df_group_Alphabet['team_Name']==team_B,'win'] += 1
        df_group_Alphabet.loc[df_group_Alphabet['team_Name']==team_B,'point'] += 3
        
        
        
        # df_group_A[(df_group_A['team_Name']=='team_A')]['win'] += 1 
    #     df_group_A[df_group_A['team_Name']=='team_A']['point'] = df_group_A[df_group_A['team_Name']=='team_A']['point'] + 3

    # else :
    #     df_group_A[df_group_A['team_Name']=='team_A']['lose'] = df_group_A[df_group_A['team_Name']=='team_A']['lose'] + 1
    #     df_group_A[df_group_A['team_Name']=='team_B']['win'] = df_group_A[df_group_A['team_Name']=='team_B']['win'] + 1 
    #     df_group_A[df_group_A['team_Name']=='team_B']['point'] = df_group_A[df_group_A['team_Name']=='team_B']['point'] + 3
   

def run_Group_Games(df_group_Alphabet):
    
    group_Teams = [team_Name for team_Name in df_group_Alphabet['team_Name']] 
    # ['Netherlands', 'Ecuador', 'Senegal', 'Qatar']
    
    group_games = list(combinations(group_Teams, 2)) 
    # [('Netherlands', 'Ecuador'), ('Netherlands', 'Senegal'), ('Netherlands', 'Qatar'), ('Ecuador', 'Senegal'), ('Ecuador', 'Qatar'), ('Senegal', 'Qatar')]
    
    
    print(f"group_games 는 {group_games}")
    print(f"group_games[0] 는 {group_games[0]}")
    
    
    for i in range(len(group_games)):
        print("2222222222")
        print(group_games[i][0])
        print(type(group_games[i][0]))
        
        run_One_Game(group_games[i][0],
                     group_games[i][1],
                     df_group_Alphabet.loc[df_group_Alphabet['team_Name']==group_games[i],'rating'],
                     df_group_Alphabet.loc[df_group_Alphabet['team_Name']==group_games[i],'rating'],
                     df_group_Alphabet)

        
# Main 문        
run_Group_Games(df_group_A)    

    
# run_One_Game('Netherlands','Ecuador',2040,1833)

# df_group_A
```
