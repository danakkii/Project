```python
import os
import pandas as pd
from PIL import Image

# Read the hotel info DataFrame
ht_info = pd.read_csv('C:/Users/user/Desktop/final_project/day/0328/web_갓다나/web2/호텔_리스트_최종.csv')

# Loop through each hotel
for index, row in ht_info.iterrows():
    # Get the folder name for the hotel
    folder_name = row['ht_id']
    
    # Get the path to the hotel image
    image_path = os.path.join("'C:/Users/user/Desktop/final_project/day/0328/web_갓다나/web2/static/img/downloads'", folder_name, "hotel.jpg")
    
    # Check if the image file exists
    if os.path.exists(image_path):
        # Open the image file
        image = Image.open(image_path)
        
        # Show the image
        image.show()
    else:
        print(f"Image file for hotel {row['ht_name']} not found.")
```
